# Atualização Bayesiana

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import beta

# Estratégia conhecida
a1, b1 = 2, 3
rv1 = beta(a1, b1)

# Nova estratégia: cenário pessimista
# a2, b2 = 0.1, 2
# rv2 = beta(a2, b2)

# Nova estratégia: cenário neutro
a2, b2 = 1, 1
rv2 = beta(a2, b2)

# Nova estratégia: cenário otimista
# a2, b2 = 8, 4
# rv2 = beta(a2, b2)

x = np.linspace(0, 1, 100)

plt.plot(x, rv1.pdf(x), label=r'Beta($\alpha=%.0f,\ \beta=%.0f$)' % (a1, b1))
plt.plot(x, rv2.pdf(x), label=r'Beta($\alpha=%.0f,\ \beta=%.0f$)' % (a2, b2))

plt.legend()

plt.show()
```

## Adicionar observações
#### Somar ao a1 os 'sucessos' da estratégia azul e ao b1 os 'insucessos' da estratégia azul
#### Somar ao a2 os 'sucessos' da estratégia laranja e ao b2 os 'insucessos' da estratégia laranja
